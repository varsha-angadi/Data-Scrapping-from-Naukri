```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/it-jobs?src=gnbjobs_homepage_srch")
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
soup
```

```python
Job_Name=soup.find_all('a',class_='title')
Job_Name
```

```python
Job_Name_list=[]
for i in Job_Name[:19]:
    Job_Name_list.append(i.text)
Job_Name_list
```

```python
Company_Name=soup.find_all('a',class_='comp-name mw-25')
Company_Name
```

```python
Company_Name_list=[]
for i in Company_Name:
    Company_Name_list.append(i.text)
Company_Name_list
```

```python
Exp_Req=soup.find_all('span',class_='expwdth')
Exp_Req
```

```python
Exp_req_List=[]
for i in Exp_Req[:19]:
    Exp_req_List.append(i.text)
Exp_req_List
```

```python
Location=soup.find_all('span',class_='locWdth')
Location
```

```python
Location_List=[]
for i in Location[:19]:
    Location_List.append(i.text)
Location_List
```

```python
data={
    'Job_Name':Job_Name_list,
    'Company_Name':Company_Name_list,
    'Exp_req':Exp_req_List,
    'Location':Location_List
}
data
```

```python
driver.quit()
```

```python
df=pd.DataFrame(data)
```

```python
df.to_csv('Job_Vacancy.csv')
```
